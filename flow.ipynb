```python
import RPi.GPIO 
import time

flow_sensor_gpio = 4

RPi.GPIO.setmode(RPi.GPIO.BCM)
RPi.GPIO.setup(flow_sensor_gpio, RPi.GPIO.IN, pull_up_down=RPi.GPIO.PUD_UP)
```

```python
count = 0
counting = True
```

```python
def countPulse(channel):
    global count
    global counting
    if counting == True:
        count = count+1

RPi.GPIO.add_event_detect(flow_sensor_gpio, RPi.GPIO.FALLING, callback=countPulse)
```

```python
def get_flow():
    global count 
    global counting
    count = 0     
    counting = True
    time.sleep(1)
    counting = False
    flow = (count / 7.5) # Pulse frequency (Hz) = 7.5Q, Q is flow rate in L/min.    
    return flow
```

```python
get_flow()
```

```python
while True:    
    flow = get_flow()
    print("The flow is: %.3f Liter/min" % (flow))
    count = 0
    time.sleep(1)
```

```python
GPIO.cleanup()
```
